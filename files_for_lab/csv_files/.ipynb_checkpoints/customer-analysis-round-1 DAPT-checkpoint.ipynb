# Perform the data cleaning operations mentioned so far in class:

# Delete the column education and the number of open complaints from the dataframe.

```python
data = data.drop(['education','number_of_open_complaints'], axis=1)
data
```

# Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type.

```python
data.dtypes
```

```python
listclv = data['customer_lifetime_value']
listclv
```

```python
listclv = data['customer_lifetime_value']
listfile = []
filereplace = []

for i in listclv:
    listfile.append(str((i)))
    
for i in listfile:
    filereplace.append(i.replace('%',''))

filereplace = pd.to_numeric(filereplace, errors='coerce')
                              
data['customer_lifetime_value'] = filereplace/100

data
```
